```python
import numpy as np

np.random.seed(42)

matrix_a = np.random.randint(0, 10, size=(3, 3))
matrix_b = np.random.randint(0, 10, size=(3, 3))

print("Matrix A:")
print(matrix_a)

print("\nMatrix B:")
print(matrix_b)

sum_matrix_operator = matrix_a + matrix_b

sum_matrix_function = np.add(matrix_a, matrix_b)

print("\nResult of Matrix Addition (A + B):")
print(sum_matrix_operator)

diff_matrix_operator = matrix_a - matrix_b

diff_matrix_function = np.subtract(matrix_a, matrix_b)

print("\nResult of Matrix Subtraction (A - B):")
print(diff_matrix_operator)

row, col = 1, 2
print(f"\nVerifying element at position [{row}, {col}]")

manual_sum = matrix_a[row, col] + matrix_b[row, col]
print(f"Manual Addition: {matrix_a[row, col]} + {matrix_b[row, col]} = {manual_sum}")

numpy_sum = sum_matrix_operator[row, col]
print(f"NumPy Addition: {numpy_sum}")

manual_diff = matrix_a[row, col] - matrix_b[row, col]
print(f"Manual Subtraction: {matrix_a[row, col]} - {matrix_b[row, col]} = {manual_diff}")

numpy_diff = diff_matrix_operator[row, col]
print(f"NumPy Subtraction: {numpy_diff}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

matrix_a = np.array([[1, 2],
                     [3, 4],
                     [5, 6]])

matrix_b = np.array([[7, 8, 9],
                     [10, 11, 12]])

print("Matrix A (3x2):")
print(matrix_a)

print("\nMatrix B (2x3):")
print(matrix_b)

result_matrix = matrix_a @ matrix_b

print("\nResult of A @ B (3x3):")
print(result_matrix)

image_matrix = np.array([[200, 150],
                         [100, 50]])

transformation_matrix = np.array([[1.5, 0],
                                  [0, 1.5]])

transformed_image = image_matrix @ transformation_matrix

print("\nOriginal 'Image' Matrix (2x2):")
print(image_matrix)
print("\nTransformation 'Filter' Matrix (2x2):")
print(transformation_matrix)
print("\nTransformed 'Image' Matrix (2x2):")
print(transformed_image)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Matrix Multiplication Visualization')

axs[0].imshow(matrix_a, cmap='viridis')
axs[0].set_title("Matrix A (3x2)")

axs[1].imshow(matrix_b, cmap='viridis')
axs[1].set_title("Matrix B (2x3)")

axs[2].imshow(result_matrix, cmap='viridis')
axs[2].set_title("Result (3x3)")

plt.show()
```
